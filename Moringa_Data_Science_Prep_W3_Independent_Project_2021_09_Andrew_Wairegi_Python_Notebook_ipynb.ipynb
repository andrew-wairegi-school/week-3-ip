It appears that they're are multiple ID's of the cells. This means that they are most likely cell tower ID's. Meaning they do not correlate to the user's cellphone ID

```python
cell_ids_group = info_df_all[["value", "cell_id"]].groupby("cell_id")
cell_ids_values = (cell_ids_group.sum()).sort_values(by = "value", ascending = False)

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
  print(cell_ids_values)
```

It looks like there are a lot of cell towers which are being used that have a lot of value. But to determine which one's to cutoff. We also need to take into account the number of users for each cell tower

```python
# Let's first figure out a cutoff point for the amount of value
# For each cell tower.
```

```python
# SMS_rate = 25sh
# Call_rate = 28sh/minute

# SMS = 8 SMSs
# Call = 2 minute * 7 = 14 minutes

# Anything above this amount would be fine
# For a cell tower over 3 days

(25 * 8) + (56 * 7)

# Anything over 600 in value
# Is ok. Cause it could be a rising cell tower.
# in population. Meaning there may be more people using it
# Over time. As the population grows.
```

We need to determine the number of times a cell tower is used

```python
cell_ids_count = cell_ids_group.count()
cell_ids_count.columns = ["Count"]
cell_ids_count = cell_ids_count.sort_values(by = "Count", ascending = False)

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
  print(cell_ids_count[cell_ids_count["Count"] >= 10])
```

A good number looks like 15. For expansion

Now we can decide on the value and amount of users to look for.

1. Join the tables

```python
cell_ids_join = cell_ids_values.join(cell_ids_count)
cell_ids_join
```

```python
# Check if any are null
cell_ids_join.isnull().sum()
```

```python
search_results = cell_ids_join[(cell_ids_join["value"] > 600) & (cell_ids_join["Count"] > 15)]
search_results
```
